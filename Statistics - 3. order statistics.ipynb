```python
# import seaborn (statistics plotting package)
import seaborn as sns
# settings for seaborn plotting style
sns.set(color_codes=True)
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(5,5)})
```

```python
# set number of draws
n = 5000
```

## Uniform distribution

```python
# import uniform distribution
from scipy.stats import uniform
```

```python
# random numbers from uniform distribution
start = 0
width = 1
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
```

```python
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True, # plot a gaussian kernel density estimate
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

### Simulate order statistics from uniform distribution

```python
# set bounds for uniform distribution
start = 0
width = 1
# set number of order statistics to draw
sample_n = 1000
# set size of random number to draw from each time (number of bidders)
N = 2
# set the order statistic (e.g. 2 for 2nd order statistics)
order = 2
```

```python
# initiate an empty array to store generated results
order_stat = []

# draw order statistics repeatedly
for i in range(0, sample_n):
    # generate an array of random numbers from the specified random distribution
    data_random = uniform.rvs(size=N, loc = start, scale=width)
    # sort and select the n-th highest value from the generated numbers
        # np.partition is a fast sort function that partially sort the list 
        # -order selects the order-th element from the end of the list
    order_stat_value = np.partition(data_random, -order)[-order]
    # store the n-th highest value in a list
    order_stat.append(order_stat_value)

# print the generated values of order statistics
# order_stat
```

```python
# calculate sample mean of the order statistics
order_stat_mean = np.mean(order_stat)
# print the mean value
print("mean = " + str(order_stat_mean) )

# plot the distribution of the order statistics
order_stat_plot = sns.distplot(order_stat,
                  bins=100,
                  kde=True, # plot a gaussian kernel density estimate
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
# add vertical line to mark the sample mean
order_stat_plot.axvline(order_stat_mean, color='orange', linestyle='--')
# add legend in graph
order_stat_plot.legend({'Mean':order_stat_mean})
# add labels to axis
order_stat_plot.set(xlabel='2nd order statistics from uniform distribution ', ylabel='Frequency')
```

## Normal distribution

```python
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal = norm.rvs(size=n,loc=0,scale=1)
```

```python
ax = sns.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

## Exponential distribution

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=n)
```

```python
ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```
